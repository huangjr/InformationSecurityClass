```python
# extended gcd to find multiplicative inverse
# gcd(a,b)=ax+by, I want to keep the record of x,y
def gcdExtended(a, b, x, y): 
    # Base Case 
    if a == 0 :  
        x = 0
        y = 1
        return (b,x,y)   
    x1 = 1
    y1 = 1 # To store results of recursive call 
    gcd,gx,gy = gcdExtended(b%a, a, x1, y1) 
    # Update x and y using results of recursive 
    # call 
    x = int(y1 - (b/a) * x1 )
    y = int(x1) 
    gx=gx+x
    gy=gy+y
    return (gcd,gx,gy) 
x = 1
y = 1
a = 100
b = 40
g,gx,gy = gcdExtended(a, b, x, y) 
print("gcd(", a , "," , b, ") = ", g) 
print("bezout codfficient=",gx,",",gy)
```

```python
def modInverse(a, m) : 
    m0 = m 
    y = 0
    x = 1
  
    if (m == 1) : 
        return 0
  
    while (a > 1) : 
        q = a // m 
  
        t = m 
  
        m = a % m 
        a = t 
        t = y 
        
        y = x - q * y 
        x = t 
    if (x < 0) : 
        x = x + m0 
  
    return x 
```
